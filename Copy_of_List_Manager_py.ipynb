```python
FILE_NAME = "tasks.txt"


def load_list():
    """Load items from the file. Return an empty list if the file does not exist."""
    try:
        with open(FILE_NAME, "r") as file:
            items = []

            for line in file:
                item = line.strip()

                if item:
                    items.append(item)

            return items

    except FileNotFoundError:
        print("No saved list was found. A new list will be created.")
        return []


def save_list(items):
    """Save all list items to the file."""
    try:
        with open(FILE_NAME, "w") as file:
            for item in items:
                file.write(item + "\n")

        print("List saved successfully.")

    except OSError:
        print("The list could not be saved.")


def display_list(items):
    """Display the numbered list."""
    print("\nYour List:")

    if not items:
        print("The list is currently empty.")
        return

    for number, item in enumerate(items, start=1):
        print(f"{number}. {item}")


def add_item(items):
    """Add a new item to the list."""
    item = input("Enter the new item: ").strip()

    if not item:
        print("The item cannot be blank.")
        return

    items.append(item)
    print(f'"{item}" was added.')


def get_item_number(items, prompt):
    """Ask the user for a valid list position."""
    if not items:
        print("The list is empty.")
        return None

    try:
        item_number = int(input(prompt))

        if item_number < 1 or item_number > len(items):
            print("That item number is not in the list.")
            return None

        return item_number - 1

    except ValueError:
        print("Please enter a valid whole number.")
        return None


def remove_item(items):
    """Remove an item from the list."""
    display_list(items)

    index = get_item_number(items, "Enter the number of the item to remove: ")

    if index is not None:
        removed_item = items.pop(index)
        print(f'"{removed_item}" was removed.')


def edit_item(items):
    """Replace an existing item with new text."""
    display_list(items)

    index = get_item_number(items, "Enter the number of the item to edit: ")

    if index is None:
        return

    new_item = input("Enter the updated item: ").strip()

    if not new_item:
        print("The updated item cannot be blank.")
        return

    old_item = items[index]
    items[index] = new_item

    print(f'"{old_item}" was changed to "{new_item}".')


def move_item(items):
    """Move an item to another position in the list."""
    display_list(items)

    current_index = get_item_number(
        items,
        "Enter the number of the item to move: "
    )

    if current_index is None:
        return

    try:
        new_position = int(
            input(f"Enter its new position from 1 to {len(items)}: ")
        )

        if new_position < 1 or new_position > len(items):
            print("That position is not valid.")
            return

        item = items.pop(current_index)
        items.insert(new_position - 1, item)

        print(f'"{item}" was moved to position {new_position}.')

    except ValueError:
        print("Please enter a valid whole number.")


def list_manager():
    """Run the list manager until the user chooses to quit."""
    items = load_list()

    print("Welcome to the List Manager!")

    while True:
        print("\nChoose an option:")
        print("1. View list")
        print("2. Add item")
        print("3. Remove item")
        print("4. Edit item")
        print("5. Move item")
        print("6. Save list")
        print("7. Save and quit")

        choice = input("\nEnter your choice: ").strip()

        if choice == "1":
            display_list(items)

        elif choice == "2":
            add_item(items)

        elif choice == "3":
            remove_item(items)

        elif choice == "4":
            edit_item(items)

        elif choice == "5":
            move_item(items)

        elif choice == "6":
            save_list(items)

        elif choice == "7":
            save_list(items)
            print("Thank you for using the List Manager!")
            break

        else:
            print("Invalid choice. Please select a number from 1 to 7.")


list_manager()
```
