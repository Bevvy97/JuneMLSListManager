```python
def get_number(prompt):
    """Ask for a number and keep trying until valid input is entered."""
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Invalid entry. Please enter a number.")


def calculate(num1, num2, operation):
    """Perform the selected calculation."""
    if operation == "+":
        return num1 + num2

    elif operation == "-":
        return num1 - num2

    elif operation == "*":
        return num1 * num2

    elif operation == "/":
        if num2 == 0:
            return "Error: You cannot divide by zero."
        return num1 / num2

    elif operation == "**":
        return num1 ** num2

    elif operation == "%":
        if num2 == 0:
            return "Error: You cannot use zero as the divisor."
        return num1 % num2

    else:
        return "Error: Invalid operation."


def calculator():
    print("Welcome to the Python Calculator!")

    while True:
        print("\nChoose an operation:")
        print("+  Addition")
        print("-  Subtraction")
        print("*  Multiplication")
        print("/  Division")
        print("** Exponent")
        print("%  Remainder")
        print("q  Quit")

        operation = input("\nEnter your choice: ").strip().lower()

        if operation == "q":
            print("Thank you for using the calculator!")
            break

        valid_operations = ["+", "-", "*", "/", "**", "%"]

        if operation not in valid_operations:
            print("Invalid operation. Please choose from the menu.")
            continue

        num1 = get_number("Enter the first number: ")
        num2 = get_number("Enter the second number: ")

        result = calculate(num1, num2, operation)
        print(f"Result: {result}")


calculator()
```
